```python
import pandas as pd
import numpy as np
```

```python
df_True  = pd.read_csv("archive/True.csv")
df_True
```

```python
df_Fake = pd.read_csv("archive/Fake.csv")
df_Fake
```

```python
df1= df_True.drop(['subject','date'], axis = 1)
df1['label'] = 'REAL'
df1
```

```python
df2= df_Fake.drop(['subject','date'], axis = 1)
df2['label']  = 'FAKE'
df2
```

```python
full_data = pd.concat([df1,df2])
full_data.reset_index()
```

full_data.to_csv('news_kaggle.csv')
